A semi-positive 2x2 symmetric matrices is given by 
$ \left| \begin{array}
x & z \\ z & y 
\end{array} \right|$
with $x \geq 0$, $y \geq 0$ and $|z| \leq \sqrt{x y}$.

The space of rank 1 matrices consists of the boundary surface $|z| = \sqrt{x y}$ without $(0,0,0)$, which is homeomorphic to a 1-holed 2d-plane.

It does not quite make sense to call a function concave for the whole space of rank 1 matrices in the embedded space ($\mathbb{R}^3$ or $\mathbb{R}^4$). However it makes sense if we restrict it to any convex subset.

```python
from mpl_toolkits import mplot3d 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x0 = np.linspace(0, 1, 32)

x = np.outer(x0, np.ones(32)) 
y = x.copy().T 
z = np.sqrt(x*y)

fig = plt.figure(figsize =(10, 10)) 
ax = plt.axes(projection ='3d') 

ax.plot_surface(x, y, z) 
ax.plot_surface(x, y, -z)

l_1 = 0.5
ax.plot(x0, l_1*x0, np.sqrt(l_1)*x0)

l_2 = 0.9
ax.plot(x0, l_2*x0, np.sqrt(l_2)*x0)

plt.show() 
```
